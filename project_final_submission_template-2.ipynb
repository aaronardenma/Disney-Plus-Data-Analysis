```python
from cs103 import *
from typing import Optional, NamedTuple, List
from enum import Enum
import csv

ContentType = Enum('ContentType', ['Movie',
                                   'TV_Show'])
#interp. the type of Disney Plus content, Movie or TV Show
#examples are redundant for Enumerations

#template baseed on Enumeration, 2 cases
@typecheck
def fn_for_content_type(ct: ContentType) -> ...:
    if ct == ContentType.Movie:
        return ...
    elif ct == ContentType.TV_Show:
        return ...

ContentRating = Enum('ContentRating', ['G',
                                       'PG',
                                       'PG_13',
                                       'TV_14',
                                       'TV_G',
                                       'TV_PG',
                                       'TV_Y',
                                       'TV_Y7',
                                       'TV_Y7_FV',
                                       'No_Rating',
                                       ])
#interp. the type of Disney Plus ratings: G, PG, PG-13, TV-14, TV-G, TV-PG, TV-Y, TV-Y7, TV-Y7-FV, NoRating, and no input in rating
#examples are redundant for Enumerations

#template based on Enumeration, 10 cases 
@typecheck
def fn_for_content_rating(cr: ContentRating) -> ...:
    if cr == ContentRating.G:
        return ...
    elif cr == ContentRating.PG:
        return ...
    elif cr == ContentRating.PG_13:
        return ...
    elif cr == ContentRating.TV_14:
        return ...
    elif cr == ContentRating.TV_G:
        return ...
    elif cr == ContentRating.TV_PG:
        return ...
    elif cr == ContentRating.TV_Y:
        return ...
    elif cr == ContentRating.TV_Y7:
        return ...
    elif cr == ContentRating.TV_Y7_FV:
        return ...
    elif cr == ContentRating.No_Rating:
        return ...



#Data Definition for Disney Plus Content

DPContent = NamedTuple('DPContent', [('type', ContentType),           #Movie, TV Show
                                 ('release_year', int),       #in range 1900-2022
                                 ('rating', ContentRating)])            #G, PG, PG-13, TV-14, TV-G, TV-PG, TV-Y, 
                                                              #TV-Y7, TV-Y7-FV, None

#interp. disney plus content data from csv data. includes the type of content (Movie/TV Show), release year and
#rating: G, PG, PG-13, rating, TV-14, TV-G, TV-PG, TV-Y, TV-Y7, TV-Y7-FV, None

DP1 = DPContent(ContentType.Movie, 2016, ContentRating.TV_G)
DP2 = DPContent(ContentType.TV_Show, 2021, ContentRating.No_Rating)
DP3 = DPContent(ContentType.TV_Show, 2019, ContentRating.TV_14)
DP4 = DPContent(ContentType.Movie, 2021, ContentRating.PG_13)
DP5 = DPContent(ContentType.TV_Show, 2019, ContentRating.TV_PG)
DP6 = DPContent(ContentType.Movie, 2008, ContentRating.G)
DP7 = DPContent(ContentType.TV_Show, 2018, ContentRating.TV_Y7)
DP8 = DPContent(ContentType.Movie, 1993, ContentRating.PG)
DP9 = DPContent(ContentType.Movie, 2021, ContentRating.TV_Y7_FV)
DP10 = DPContent(ContentType.Movie, 2021, ContentRating.TV_Y)

#template based on compound and reference rule
@typecheck
def fn_for_dpc(dpc: DPContent) -> ...:
    return ...(fn_for_content_type(dpc.type),
               dpc.release_year,
               fn_for_content_rating(dpc.rating))

# Data Definition for List of Disney Plus Content

# List[DPContent]
# interp. Represents a list of disney plus content data read from a file provided by Disney Plus

LOC0 = []
LOC1 = [DP1]
LOC2 = [DP1, DP2, DP3]
LOC3 = [DP1, DP2, DPContent(ContentType.Movie, 2021, ContentRating.TV_G)]
LOC4 = [DPContent(ContentType.Movie, 2021, ContentRating.TV_PG),
        DPContent(ContentType.TV_Show, 2017, ContentRating.TV_Y7)]

#template based on arbitrary sized and reference rule
@typecheck
def fn_for_lodpc(lodpc: List[DPContent]) -> ...:
    #accumulator description
    acc = ... #type: ...
    
    for dpc in lodpc:
        acc = ...(acc, fn_for_lodpc(dpc))
    return ...(acc)
```

```python
#row 1 -> type
#row 7 -> release year
#row 8 -> rating

###########
# Functions

@typecheck
def parse_content_type(dpc: str) -> ContentType:
    """
    Given a string which must be one of "Movie" or "TV Show",
    return the corresponding ContentType value
    """
    #return ContentType.Movie # stub
    
    #template based on simple atomic
    #return ...(dpc)
    if dpc == "Movie":
        return ContentType.Movie
    elif dpc == "TV Show":
        return ContentType.TV_Show

@typecheck
def parse_content_rating(dpc: str) -> ContentRating:
    """
    Given a string which must be one of 'G','PG','PG_13','TV_14',
    'TV_G','TV_PG','TV_Y','TV_Y7','TV_Y7_FV','None'
    return the corresponding ContentRating value
    """
    #return ContentRating.G #stub
    
    #template based on simple atomic
    #return ...(dpc)
    if dpc == "G":
        return ContentRating.G
    elif dpc == "PG":
        return ContentRating.PG
    elif dpc == "PG-13":
        return ContentRating.PG_13
    elif dpc == "TV-14":
        return ContentRating.TV_14
    elif dpc == "TV-G":
        return ContentRating.TV_G
    elif dpc == "TV-PG":
        return ContentRating.TV_PG
    elif dpc == "TV-Y":
        return ContentRating.TV_Y
    elif dpc == "TV-Y7":
        return ContentRating.TV_Y7
    elif dpc == "TV-Y7-FV":
        return ContentRating.TV_Y7_FV
    elif dpc == "None":
        return ContentRating.No_Rating

    
@typecheck
def is_reliable(columns: List[str]) -> bool:
    '''
    Returns True if the given list of strings columns is reliable i.e. the data at
    columns[8] is not empty ''. False otherwise.
    '''
    # return True # stub
    # Tempalte treats columns as atomic and uses indexing instead
    # return ...(columns) # Template
    
    return columns[8] != ""
    
@typecheck
def read(filename: str) -> List[DPContent]:
    """    
    reads information from the specified file and returns Disney Plus Content 
    """
    #return []  #stub
    # Template from HtDAP
    # lodpc contains the result so far
    lodpc = [] # type: List[DPContent]

    with open(filename) as csvfile:
        
        reader = csv.reader(csvfile)
        next(reader) # skip header line

        for row in reader:
            if is_reliable(row):
                dpc = DPContent(parse_content_type(row[1]), parse_int(row[7]), parse_content_rating(row[8]))
                lodpc.append(dpc)
    return lodpc

# Begin testing
start_testing()

# Examples and tests for parse_content_type
expect(parse_content_type("Movie"), ContentType.Movie)
expect(parse_content_type("TV Show"), ContentType.TV_Show)

# Examples and tests for parse_content_rating
expect(parse_content_rating("G"), ContentRating.G)
expect(parse_content_rating("PG"), ContentRating.PG)
expect(parse_content_rating("PG-13"), ContentRating.PG_13)
expect(parse_content_rating("TV-14"), ContentRating.TV_14)
expect(parse_content_rating("TV-G"), ContentRating.TV_G)
expect(parse_content_rating("TV-PG"), ContentRating.TV_PG)
expect(parse_content_rating("TV-Y"), ContentRating.TV_Y)
expect(parse_content_rating("TV-Y7"), ContentRating.TV_Y7)
expect(parse_content_rating("TV-Y7-FV"), ContentRating.TV_Y7_FV)
expect(parse_content_rating("None"), ContentRating.No_Rating)



# Examples and tests for is_reliable
expect(is_reliable(['s1', 'Movie', 'Duck the Halls: A Mickey Mouse Christmas Special',
                    "Alonso Ramirez Ramos, Dave Wasson","Chris Diamantopoulos, Tony Anselmo, Tress MacNeille, Bill Farmer, Russi Taylor, Corey Burton", 
                    '', "November 26, 2021", '2016','TV-G', '23 min', "Animation, Family",
                    'Join Mickey and the gang as they duck the halls!']), True)
expect(is_reliable(['s5', 'TV Show', 'The Beatles: Get Back', '', "John Lennon, Paul McCartney, George Harrison, Ringo Starr",'',"November 25, 2021", '2021', '', '1 Season',"Docuseries, Historical, Music", 'A three-part documentary from Peter Jackson capturing a moment in music history with The Beatles.']), False)

# Examples and tests for read
expect(read('testfile_empty.csv'), [])
expect(read('testfile_small.csv'), [DPContent(ContentType.Movie, 2016, ContentRating.TV_G), 
                                    DPContent(ContentType.Movie, 1988, ContentRating.PG)])

# show testing summary
summary()
```

```python
# Filter function to sort Rating in Groupings of G/TV-G, PG/TV-PG, and PG-13/TV-14
@typecheck
def filter_G(dpc: DPContent) -> bool:
    """
    returns True if dpc.rating matches with rating
    """
    return dpc.rating == ContentRating.G 

@typecheck
def filter_TV_G(dpc: DPContent) -> bool:
    """
    returns True if dpc.rating matches with rating
    """
    return dpc.rating == ContentRating.TV_G 

@typecheck
def filter_PG(dpc: DPContent) -> bool:
    """
    returns True if dpc.rating matches with rating
    """
    return dpc.rating == ContentRating.PG 

@typecheck
def filter_TV_PG(dpc: DPContent) -> bool:
    """
    returns True if dpc.rating matches with rating
    """
    return dpc.rating == ContentRating.TV_PG 

@typecheck
def filter_PG_13(dpc: DPContent) -> bool:
    """
    returns True if dpc.rating matches with rating
    """
    return dpc.rating == ContentRating.PG_13

@typecheck
def filter_TV_14(dpc: DPContent) -> bool:
    """
    returns True if dpc.rating matches with rating
    """
    return dpc.rating == ContentRating.TV_14

# Begin testing
start_testing()

# Examples and tests for filter_G
expect(filter_G(DP1), False)
expect(filter_G(DP6), True)

# Examples and tests for filter_TV_G
expect(filter_TV_G(DP1), True)
expect(filter_TV_G(DP2), False)

# Examples and tests for filter_PG
expect(filter_PG(DP8), True)
expect(filter_PG(DP3), False)

# Examples and tests for filter_TV_PG
expect(filter_TV_PG(DP3), False)
expect(filter_TV_PG(DP5), True)

# Examples and tests for filter_PG_13
expect(filter_PG_13(DP4), True)
expect(filter_PG_13(DP10), False)

# Examples and tests for filter_TV_14
expect(filter_TV_14(DP3), True)
expect(filter_TV_14(DP9), False)

# show testing summary
summary()
```

```python
@typecheck
def distinct_ratings(lodpc: List[DPContent], movierating: ContentRating, tvrating: ContentRating) -> int:
    # accumulator stores dpc.ratings of DPContent in list
    acc = [] # type: List[dpc.rating]
    
    if movierating == ContentRating.G and tvrating == ContentRating.TV_G:
        for dpc in lodpc:
            if filter_G(dpc) or filter_TV_G(dpc):
                acc.append(dpc.rating)
        return len(set(acc))
    elif movierating == ContentRating.PG and tvrating == ContentRating.TV_PG:
        for dpc in lodpc:
            if filter_PG(dpc) or filter_TV_PG(dpc):
                acc.append(dpc.rating)
        return len(set(acc))
    elif movierating == ContentRating.PG_13 and tvrating == ContentRating.TV_14:
        for dpc in lodpc:
            if filter_PG_13(dpc) or filter_TV_14(dpc):
                acc.append(dpc.rating)
        return len(set(acc))
    else:
        return 0

@typecheck
def avg_number_ratings(lodpc: List[DPContent]) -> List[float]:
    if len(lodpc) == 0:
        return [0, 0, 0]
    # accumulator stores all DPContent values that are ContentRating.G and ContentRating.TV_G
    gtvg = [] # type: List[DPContent]
    
    # accumulator stores all DPContent values that are ContentRating.PG and ContentRating.TV_PG
    pgtvpg = [] #type: List[DPContent]
    
    # accumulator stores all DPContent values that are ContentRating.PG_13 and ContentRating.TV_PG_14
    pg13tv14 = [] #type: List[DPContent]
    
    # accumulator stores the average number of content in categories of G/TV-G, PG/TV-PG
    # and PG-13/TV-14 in the following order in a list
    avg = [] # type: List[float]
    
    for dpc in lodpc:
        if filter_G(dpc) or filter_TV_G(dpc):
            gtvg.append(dpc)
            
    if gtvg == [] or distinct_ratings(gtvg, ContentRating.G, ContentRating.TV_G) == 0:
        average_gtvg = 0
        avg.append(average_gtvg)
    else:
        avg.append(len(gtvg)/distinct_ratings(gtvg, ContentRating.G, ContentRating.TV_G))
    
    for dpc in lodpc:
        if filter_PG(dpc) or filter_TV_PG(dpc):
            pgtvpg.append(dpc)
                
    if pgtvpg == [] or distinct_ratings(pgtvpg, ContentRating.PG, ContentRating.TV_PG) == 0:
        average_pgtvpg = 0
        avg.append(average_pgtvpg)
    else:
        avg.append(len(pgtvpg)/distinct_ratings(pgtvpg, ContentRating.PG, ContentRating.TV_PG))
    
    for dpc in lodpc:
        if filter_PG_13(dpc) or filter_TV_14(dpc):
            pg13tv14.append(dpc)
        
    if pg13tv14 == [] or distinct_ratings(pg13tv14, ContentRating.PG_13, ContentRating.TV_14) == 0:
        average_pg13tv14 = 0
        avg.append(average_pg13tv14)
        
    else:
        avg.append(len(pg13tv14)/distinct_ratings(pg13tv14, ContentRating.PG_13, ContentRating.TV_14))
    return avg

 

# Begin testing
start_testing()

# Examples and tests for distinct_ratings
expect(distinct_ratings(LOC0, ContentRating.G, ContentRating.TV_G), 0)
expect(distinct_ratings(LOC2, ContentRating.G, ContentRating.TV_G), 1)
expect(distinct_ratings([DPContent(ContentType.Movie, 2021, ContentRating.G),
                   DPContent(ContentType.TV_Show, 2021, ContentRating.TV_G),
                   DPContent(ContentType.Movie, 2021, ContentRating.TV_PG),
                   DPContent(ContentType.TV_Show, 2021, ContentRating.TV_G)], ContentRating.G, ContentRating.TV_G), 2)

# Examples and tests for avg_num_rating
expect(avg_number_ratings(LOC0), [0, 0, 0])
expect(avg_number_ratings(LOC2), [1, 0, 1])
expect(avg_number_ratings([DPContent(ContentType.Movie, 1990, ContentRating.G),
                   DPContent(ContentType.TV_Show, 1990, ContentRating.TV_G),
                   DPContent(ContentType.Movie, 1990, ContentRating.TV_PG),
                   DPContent(ContentType.TV_Show, 1990, ContentRating.TV_G)]), [1.5, 1, 0])
                                                                                                                          

# show testing summary
summary()
```
